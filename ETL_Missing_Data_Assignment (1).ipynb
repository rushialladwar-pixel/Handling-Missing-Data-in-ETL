## SECTION A – THEORETICAL ANSWERS

### Q1. Common reasons for missing data in ETL
- Data entry errors
- System failures or data corruption
- Optional fields not filled by users
- Data integration issues from multiple sources
- Delayed data collection

### Q2. Why blindly deleting rows is bad?
- Loss of valuable information
- Bias in dataset
- Reduced sample size leading to poor analysis

### Q3. Listwise vs Column deletion
- **Listwise deletion:** Remove rows with missing values  
  *Use when missing data is small and random*
- **Column deletion:** Remove entire column  
  *Use when column has too many missing values (>50%)*

### Q4. Median vs Mean
Median is better for skewed data because it is not affected by extreme values.

### Q5. Forward Fill
- Fills missing values using previous row value  
- Useful in **time-series or sequential data**

### Q6. Why flag before imputation?
- Keeps track of original missing values
- Helps in better modeling and analysis

### Q7. Business insight from missing income
- Customers hiding income may indicate privacy concerns
- Could represent a specific segment (e.g., freelancers)
- Helps in targeted marketing strategies


```

import pandas as pd
import numpy as np

# Creating dataset
data = {
    "Customer_ID": [101,102,103,104,105,106,107,108],
    "Name": ["Rahul Mehta","Anjali Rao","Suresh Iyer","Neha Singh","Amit Verma","Karan Shah","Pooja Das","Riya Kapoor"],
    "City": ["Mumbai","Bengaluru","Chennai","Delhi","Pune","Ahmedabad","Kolkata","Jaipur"],
    "Monthly_Sales": [12000,np.nan,15000,np.nan,18000,np.nan,14000,16000],
    "Income": [65000,np.nan,72000,np.nan,58000,61000,np.nan,69000],
    "Region": ["West","South","South","North",np.nan,"West","East","North"]
}

df = pd.DataFrame(data)
df
```

## Q8. Listwise Deletion

```

# Identify rows with missing Region
missing_region = df[df['Region'].isna()]
missing_region
```

```

# Remove rows
df_q8 = df.dropna(subset=['Region'])
df_q8
```

```

# Count lost records
lost_records = len(df) - len(df_q8)
lost_records
```

## Q9. Forward Fill

```

# Before
df[['Customer_ID','Monthly_Sales']]
```

```

# Apply forward fill
df['Monthly_Sales_Filled'] = df['Monthly_Sales'].fillna(method='ffill')

df[['Customer_ID','Monthly_Sales','Monthly_Sales_Filled']]
```


Forward fill is suitable because sales data often follows a sequence where previous values can approximate missing ones.


## Q10. Flagging Missing Data

```

# Create flag column
df['Income_Missing_Flag'] = df['Income'].isna().astype(int)

df
```

```

# Count missing income
missing_income_count = df['Income_Missing_Flag'].sum()
missing_income_count
```
